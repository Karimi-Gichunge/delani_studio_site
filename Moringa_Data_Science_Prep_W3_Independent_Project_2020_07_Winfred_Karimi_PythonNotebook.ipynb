```
#import the pandas library
#
import pandas as pd
#import the Numpy library
import numpy as np
```

```
#Load our csv files and create a dataframe with each of them
#load dataset 1 as dataframe "day1"
day1 = pd.read_csv('Telcom_dataset1.csv')
day1.head()
#Here we will drop all columns that are not necessary
to_drop = ['DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'SITE_ID']
day1.drop(to_drop, inplace = True, axis = 1)
#Read in the first 5 lines of the dataframe "day1"
day1.head()
```

```
#load dataset 2 as "day2"
day2 = pd.read_csv('Telcom_dataset2.csv')
day2.head()
#We will then drop all the attributes that are not important
to_drop = ['DW_A_NUMBER', 'DW_B_NUMBER', 'COUNTRY_A', 'COUNTRY_B', 'SITE_ID']
day2.drop(to_drop, inplace = True, axis = 1)
#Read in the first 5 entrioes of dataframe "day2"
day2.head() 
```

```
#load dataset 3 as "day3"
day3 = pd.read_csv('Telcom_dataset3.csv')
day3.head()
#We drop all the attributes that are not deemed important in the study
to_drop = ['DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'SIET_ID']
day3.drop(to_drop, inplace = True, axis = 1)
#Read in the first 5 entries of the dataframe "day3"
day3.head() 
```

```
#We will also need to load the cells_geo.csv dataset
geo = pd.read_csv('cells_geo.csv', delimiter=';')
geo.head()
#The same process of dropping non_important columns is carried out here too
to_drop = ['LOCALISATION', 'DECOUPZONE', 'ZONENAME', 'LONGITUDE', 'LATITUDE', 'REGION', 'AREA', 'SITE_CODE']
geo.drop(to_drop, inplace = True, axis = 1)
#Read in the first 5 entries of dataframe "geo"
geo.head()
```

```
day1
#Correcting column name "PRODUTC" to "PRODUCT" in dataframe day1
day1 = day1.rename(columns={'PRODUTC':'PRODUCT', 'DATETIME':'DATE_TIME'})
day1.head()
```

```
#Change the column name "CELLID" in day3 to "CELL_ID"
day3 = day3.rename(columns={'CELLID':'CELL_ID'})
day3.head()
```

```
#Here we concatenate the dataframe day1,day2,day3 vertically

dfs = [day1, day2, day3,]
#Read in the first 5 lines of the new dataframe "merged"
merged = pd.concat(dfs)
merged.head()
 
```

```
#Remove all rows where cell_id is #NAME?
merged[merged.CELL_ID != '#NAME?']
```

```
#Remove all NaN rows from the dataframe "geo"
geo = geo[geo['VILLES'].notna()]
#Read in the last 5 entries of dataframe "geo"
geo.tail()
```

```
merged.reset_index(inplace=True)
merged
```

```
#Here, the "geo" dataframe is joined to the complete data from the 3 days
all = pd.merge(geo, merged, on='CELL_ID', how='outer')
all.tail()
```

```
#Extract a subset of all the data that contains only the cities, Products and Date/Time
subset = all[['VILLES_x', 'PRODUCT', 'DATE_TIME']]

subset.tail()
```

```
#At this point, one can now start analysing the data
#1. We will look at which cities were mostly used in the three days
x= all.groupby('VILLES_x', sort = True).size()
x.nlargest(10)
```
